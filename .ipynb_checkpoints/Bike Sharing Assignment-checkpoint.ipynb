### Importing and Understanding Data

```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data = pd.read_csv('day.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
# We have no column with null or NA values and all data types are correct
```

```python
def season_name(x):
    return 'spring' if x==1 else 'summer' if x==2 else 'fall' if x==3 else 'winter' 
data.season = data.season.apply(season_name)
```

```python
def weather_name(x):
    return 'weather_cat_1' if x==1 else 'weather_cat_2' if x==2 else 'weather_cat_3' if x==3 else 'weather_cat_4' 
data.weathersit = data.weathersit.apply(weather_name)
```

# Data Preparation

```python
# As instant column does not provide any value, we can drop it
data.drop(columns=['instant'],inplace=True)
```

```python
data.head()
```

```python
weather_status = pd.get_dummies(data['weathersit'], drop_first = True)
data = pd.concat([data, weather_status], axis = 1)
data.drop(['weathersit'], axis = 1, inplace = True)
data.head()
```

```python
season_status = pd.get_dummies(data['season'], drop_first = True)
data = pd.concat([data, season_status], axis = 1)
data.drop(['season'], axis = 1, inplace = True)
data.head()
```

We can extract the date from `dteday` and remove dteday

```python
data['date'] = pd.DatetimeIndex(data.dteday).day
data.drop(['dteday'],axis=1, inplace=True)
```

```python
data.head()
```

## Step 4: Splitting the Data into Training and Testing Sets

```python
df_train, df_test = train_test_split(data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
print(df_test.shape)
print(df_train.shape)
```

### Rescaling the Features 


```python
scaler = MinMaxScaler()
```

```python
num_vars = ['temp', 'atemp', 'atemp', 'windspeed', 'casual','registered','cnt']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train[num_vars].describe()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## Step 5: Building a linear model


```python
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

print(lr_1.summary())

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
print(vif)
```

Looks like `weather_cat_2` has the highest p-value. So we will remove that from our model

```python
X = X_train.drop('weather_cat_2', 1,)
X_train_lm = sm.add_constant(X)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

print(lr_1.summary())

vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
print(vif)
```

```python
def remove_feature_and_train_model(feature):
    global X
    global lr_1
    global X_train_lm
    X = X.drop(feature, 1,)
    X_train_lm = sm.add_constant(X)

    lr_1 = sm.OLS(y_train, X_train_lm).fit()

    print(lr_1.summary())

    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```

## Step 7: Residual Analysis of the train data


```python
y_train_price = lr_1.predict(X_train_lm)
```

```python
fig = plt.figure()
sns.distplot((y_train - y_train_price), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)  
```
